```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import keras
```

```python
def GA(N, N0, K):
    if N==32 and N0==16 and K==16:
        return np.array([5, 11]).astype(np.int8)

def GA0(N0, K0):
    if N0 == 4:
        tmp = np.array([0, 1, 2, 3]).astype(np.int8)
    elif N0 == 8:
        tmp = np.array([0, 1, 2, 4, 3, 5, 6, 7]).astype(np.int8)
    elif N0 == 16:
        tmp = np.array([0, 1, 2, 4, 8, 3, 5, 6, 9, 10, 12, 7, 11, 13, 14, 15]).astype(np.int8)
    return np.array([np.sort(tmp[N0-i:]) for i in K0])

def NN_param(N, K):
    model = keras.models.load_model('./model/model_' + str(N) + '_' + str(K) + '.h5')
    tmp = model.get_weights()
    weights = []
    for i in range(int(len(tmp)/2)):
        weights.append(np.append([tmp[2*i+1]], tmp[2*i], axis=0))
    return weights

def encoder(u):
    N = len(u)
    if N == 1:
        x = u
    else:
        x1 = encoder(u[:int(N/2)])
        x2 = encoder(u[int(N/2):])
        x = np.append((x1+x2)%2, x2)
    return x

def NNdecoder(y, group):
    tmp = y
    for i in NN0[group][:-1]:
        tmp = np.dot(np.append(1, tmp), i)
        tmp[tmp < 0] = 0
    tmp = np.dot(np.append(1, tmp), NN0[group][-1])
    tmp[tmp>0] = 1
    tmp[tmp<0] = 0
    return tmp

def SC_F(a, b):
    ta = np.tanh(a/2)
    tb = np.tanh(b/2)
    f = np.log((1 + ta*tb) / (1 - ta*tb))
    return f

def SC_G(a, b, s):
    g = b + (1 - 2*s) * a
    return g

def SCNNdecoder(y):
    u_pred = np.array([])
    u0 = []
    y0 = []
    
    group = 0
    y0.append(SC_F(y[:N0], y[N0:]))
    tmp = NNdecoder(y0[group], group)
    u_pred = np.append(u_pred, tmp)
    tmp2 = np.zeros(N0, dtype=np.int8)
    tmp2[free_set0[group]] = tmp
    u0.append(tmp2)
    
    group = 1
    y0.append(SC_G(y[:N0], y[N0:], encoder(u0[0])))
    tmp = NNdecoder(y0[group], group)
    u_pred = np.append(u_pred, tmp)
    tmp2 = np.zeros(N0, dtype=np.int8)
    tmp2[free_set0[group]] = tmp
    u0.append(tmp2)
    
    return u_pred
```

```python
n = 5
N = 2**n
K = 16
R = K/N

n0 = 4
N0 = 2**n0
K0 = GA(N, N0, K)
R0 = K0/N0

m = 2 ** (n-n0)

free_set0 = GA0(N0, K0)
free_set = []
for i in range(m):
    free_set = np.append(free_set, 16*i+free_set0[i])
free_set = free_set.astype(np.int8)

NN0 = []
for i in K0:
    NN0.append(NN_param(N0, i))
```

```python
ebn0_min = 0
ebn0_max = 6
points = 13
ebn0 = np.linspace(ebn0_min, ebn0_max, points)

bsum = np.zeros(points, dtype=np.float64)
be = np.zeros(points, dtype=np.float64)
ber = np.zeros(points, dtype=np.float64)
fsum = np.zeros(points, dtype=np.float64)
fe = np.zeros(points, dtype=np.float64)
fer = np.zeros(points, dtype=np.float64)

times_min = 1000
times_max = 500000
error_max = 1000
```

```python
for i in range(points):
    SNR = ebn0[i] + 10*np.log10(2*R)
    sigma = (1/10**(SNR/10))**(1/2)
    ii = 0
    
    while (fe[i] < error_max and ii < times_max) or (ii < times_min) :
#         u_info = np.random.randint(low=0, high=2, size=K, dtype=np.int8)
        u_info = np.ones(K, dtype=np.int8)
        u = np.zeros(N, dtype=np.int8)
        u[free_set] = u_info
        x = encoder(u)
        
        y = 1 - 2*x + np.random.randn(N)*sigma
        
        u_pred = SCNNdecoder(y)
        
        diff = np.sum((u_pred + u_info)%2)
        fe[i] += bool(diff)
        be[i] += diff
        ii += 1

    fsum[i] = ii
    bsum[i] = ii*K
    fer[i] = fe[i]/fsum[i]
    ber[i] = be[i]/bsum[i]
    print(i, ii, fe[i])
```

```python
plt.figure(figsize=(10,7.5))
legend = ['NN FER', 'NN BER']

plt.plot(ebn0, ber, 'c') 
plt.plot(ebn0, fer, 'c--') 

plt.legend(legend, loc=3)
plt.yscale('log')
plt.xlabel('$E_b/N_0$')
plt.ylabel('FER / BER')    
plt.grid(True, which='both', linestyle='--')
plt.show()
```
