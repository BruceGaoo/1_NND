```python
import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
model = keras.models.load_model('./result/model_16_4.h5')
L = pd.read_csv('./result/LOSS_16_4.csv')
P_NN = pd.read_csv('./result/P_NN_16_4.csv')
P_MAP = pd.read_csv('./result/P_MAP_16_4.csv')
```

```python
plt.figure(figsize=(20,10))
plt.scatter(range(len(L['loss'])), L['loss'], s=0.01)
```

```python
plt.figure(figsize=(10,7.5))
legend = ['NN FER', 'NN BER', 'MAP FER', 'MAP BER']

plt.plot(P_NN['ebn0'], P_NN['fer'], 'c') 
plt.plot(P_NN['ebn0'], P_NN['ber'], 'c--') 
plt.plot(P_MAP['ebn0_map'], P_MAP['fer_map'], 'r')
plt.plot(P_MAP['ebn0_map'], P_MAP['ber_map'], 'r--') 

plt.legend(legend, loc=3)
plt.yscale('log')
plt.xlabel('$E_b/N_0$')
plt.ylabel('FER / BER')    
plt.grid(True)
plt.show()
```
